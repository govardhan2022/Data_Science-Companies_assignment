```python
# 3. Analyze Text
# -- Load dictionaries and stop words --
positive_words = set(open("positive-words.txt").read().split())
negative_words = set(open("negative-words.txt").read().split())
#stop_words = set(open("StopWords_Auditor.txt").read().split())
stop_words = set(open("Complete_StopWords.txt").read().split())

# Process each text file
results = []
for file in os.listdir("."):
    if file.endswith(".txt"):
        with open(file,"r", encoding="utf-8", errors="ignore") as f:
            text = f.read()


        # Clean and tokenize
        tokens = nltk.word_tokenize(text.lower())
        tokens = [word for word in tokens if word not in stop_words]

        # Compute metrics
        positive_score = sum(1 for word in tokens if word in positive_words)
        negative_score = sum(-1 for word in tokens if word in negative_words)
        # ... (other metrics)

        results.append({
            "URL_ID": file.split(".")[0],
            "url": url,
            "Positive Score": positive_score,
            "Negative Score": negative_score,
            # ... (other metrics)
        })

# 4. Save Results
output_df = pd.DataFrame(results)
output_df.to_excel("Output Data Structure6.xlsx", index=False)
```
